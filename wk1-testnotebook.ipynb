```python
# a short function to determine the hourly wage based on the 
# annual wage?
def hourly(annualwage):
    hourlywage = annualwage/(40*52)
    
    
    return hourlywage
```

```python
hourly(60000)
```

```python
hourly(200000)
```

```python
million_hourly = hourly(1000000)
million_hourly
```

```python
# create a list
somesalaries = list(range(5))
somesalaries
```

```python
# run function on list, storing results
hourly_wages = []
for salary in somesalaries:
    hourly_wages.append( hourly(salary) )
hourly_wages
```
